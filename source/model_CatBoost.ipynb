# XÂY DỰNG NHÃN VÀ THIẾT LẬP MÔ HÌNH CATBOOST

```python
import numpy as np
import pandas as pd
import random
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## Đọc dữ liệu

```python
df = pd.read_csv('../data/list_way_vehicle_2.csv')
df
```

```python
data = pd.read_csv('../data/data_pre.csv')
data.datetime = pd.to_datetime(data["datetime"])
data
```

## Tiền xử lý

### Tách danh sách các node, way

```python

def Conv_str_2_list(data):
    df = data.copy()
    for vehicle in data:
        df.at[0,vehicle] = list(data.at[0,vehicle].replace("[","" ).replace("]","" ).split(", "))
        df.at[1,vehicle] = list(data.at[1,vehicle].replace("[","" ).replace("]","" ).split(", "))
    return df
df1 = Conv_str_2_list(df).T
df1
```

### Gán nhãn ngẫu nhiên cho mỗi xe

```python
list_vehicle = list(df1.index)
labels = random.sample(range(len(list_vehicle)+1), len(list_vehicle))
df1['label'] = 0

for vehicle_id in range(len(list_vehicle)):
    for x in df1[df1.index==list_vehicle[vehicle_id]].index:
        df1.at[x,'label'] = labels[vehicle_id]
df1
```

### loại bỏ các xe có ít way

```python
df1['length'] = [len(x) for x in df1[1]]
df1 = df1.drop(df1[df1['length']<3].index, axis=0).drop([0,'length'],axis=1)
df1
```

```python
df2 = pd.DataFrame(df1[1].values.tolist(), index=df1.index.tolist()).replace(np.nan, 0)
df2['label']= df1['label']
df2
```

## Đọc dữ liệu các xe đi cùng nhau đã được thu thập trước đó

```python
df_label = pd.read_csv('../data/df_together_final_001.csv')
df_label['itemsets'] = [x.replace('[','').replace(']','').replace(',','').replace("'",'').split() for x in df_label['itemsets']]
df_label
```

### Những xe đi cùng nhau sẽ được chuyển nhãn giống nhau

```python
for i in df_label['itemsets']:
    df2.at[i[1], 'label'] = df2.at[i[0], 'label']

print(df2['label']['51G14243'])
print(df2['label']['60A31556'])
```

```python

df2['vehicle'] = df2.index
df2 = df2.set_index('vehicle')
df2
```

```python
df3 = pd.DataFrame([pd.to_numeric(df2[x], downcast='integer') for x in df2.columns]).T
df3.dtypes
```

```python
data.isnull().any()
data.dtypes
```

```python
# thêm các thuộc tính khác vào dữ liệu như khoảng thời gian khỏi hành và kết thúc của mỗi xe, tốc độ trung bình
%%time
df3['hour_start'] = [data[data.vehicle == i].datetime.min().hour for i in df3.index]
df3['minute_start'] = [data[data.vehicle == i].datetime.min().minute for i in df3.index]
df3['second_start'] = [data[data.vehicle == i].datetime.min().second for i in df3.index]

df3['hour_end'] = [data[data.vehicle == i].datetime.max().hour for i in df3.index]
df3['minute_end'] = [data[data.vehicle == i].datetime.max().minute for i in df3.index]
df3['second_end'] = [data[data.vehicle == i].datetime.max().second for i in df3.index]

df3['speed'] = [np.mean(data[data.vehicle == i].speed) for i in df3.index]
df3
```

```python
cat_features_index = np.where( df3.drop('label', axis=1).dtypes != float)[0]
cat_features_index
```

## Chia tập dữ liệu train test cho label random full

```python
X_train, X_test, y_train, y_test = train_test_split( df3.drop('label', axis=1), df3['label'], test_size=1/7.0, random_state=0)
```

## Mô hình CatBoost

```python
# RMSSE
model_1_RMSE = CatBoostRegressor(iterations=1000,
                          learning_rate= 0.1,
                          depth=10,
                          loss_function= 'RMSE',
                          use_best_model=True,
                          #od_type = 'IncToDec',
                          eval_metric = 'RMSE',
                          random_seed=42)
#Poisson
model_1_Poisson = CatBoostRegressor(iterations=1000,
                          learning_rate= 0.1,
                          depth=10,
                          loss_function= 'Poisson',
                          use_best_model=True,
                          #od_type = 'IncToDec',
                          eval_metric = 'Poisson',
                          random_seed=42)
# MAE
model_1_MAE = CatBoostRegressor(iterations=1000,
                          learning_rate= 0.1,
                          depth=10,
                          loss_function= 'MAE',
                          use_best_model=True,
                          #od_type = 'IncToDec',
                          eval_metric = 'MAE',
                          random_seed=42)
# MAPE
model_1_MAPE = CatBoostRegressor(iterations=1000,
                          learning_rate= 0.01,
                          depth=10,
                          loss_function= 'MAPE',
                          use_best_model=True,
                          eval_metric = 'MAPE',
                          random_seed=42)
```

```python
%%time
model_1_RMSE.fit(X_train, y_train, cat_features=cat_features_index, eval_set=(X_test, y_test));
model_1_Poisson.fit(X_train, y_train, cat_features=cat_features_index, eval_set=(X_test, y_test));
model_1_MAE.fit(X_train, y_train, cat_features=cat_features_index, eval_set=(X_test, y_test));
model_1_MAPE.fit(X_train, y_train, cat_features=cat_features_index, eval_set=(X_test, y_test));
```

```python
model_1_RMSE.save_model('../src/save_model/catboost_model_1_RMSE.dump')
model_1_Poisson.save_model('../src/save_model/catboost_model_1_Poisson.dump')
model_1_MAE.save_model('../src/save_model/catboost_model_1_MAE.dump')
model_1_MAPE.save_model('../src/save_model/catboost_model_1_MAPE.dump')
```
